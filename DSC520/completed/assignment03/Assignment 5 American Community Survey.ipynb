```r
description <- describe(data$HSDegree)
```

```r
description$skew
```

Skewness is a measure of asymmetry. Typically, skewness with a value lower than -1 or higher than +1 are considered highly skewed. In this case, our data has a high negative skew, indicated by the tail on the left side. 

https://help.gooddata.com/doc/en/reporting-and-dashboards/maql-analytical-query-language/maql-expression-reference/aggregation-functions/statistical-functions/predictive-statistical-use-cases/normality-testing-skewness-and-kurtosis#:~:text=As%20a%20general%20rule%20of,the%20distribution%20is%20approximately%20symmetric.

```r
description$kurtosis
```

Kurtosis is a measurement of how sharp a unimodal distribution, or how long the tails are for the distribution. A value above 3 normally indicates that the distribution that the tails of a distribution are long. A normal distribution has a kurkosis of 0.

https://www.spcforexcel.com/knowledge/basic-statistics/are-skewness-and-kurtosis-useful-statistics#:~:text=The%20value%20is%20often%20compared,(more%20in%20the%20tails).

```r
# Z Scores
zscores <- (data$HSDegree-mean(data$HSDegree))/sd(data$HSDegree)
data <- cbind(data, zscores)
```

```r
head(data)
```

Z-Score is a measurement of individual points. If the ZScore of a point is above 0, that means that the point lies above the mean. If the ZScore is lower than 0, the point lies below the mean. ZScores are useful because they allow a data scientist to calculate the probability of a point inside the normal distribution. We can use p_norm and d_norm in R to plot out the probabilities and percentiles. 

https://www.simplypsychology.org/z-score.html#:~:text=A%20z%2Dscore%20describes%20the,it%20lies%20below%20the%20mean.

```r
p_norm <- pnorm(data$zscores)
d_norm <- dnorm(data$zscores)
data <- cbind(data, p_norm)
data <- cbind(data, d_norm)
head(data)
```

```r
ggplot(data = data, aes(x = HSDegree, y = d_norm)) + 
geom_line(color = "gray20") + 
ggtitle("ZScore Distribution") + 
scale_y_continuous(labels = scales::percent) + 
theme_minimal()

ggplot(data = data, aes(x = HSDegree, y = p_norm)) + 
geom_line(color = "gray20") + 
ggtitle("ZScore Distribution: Cumulative Probability") + 
scale_y_continuous(limits = 0:1,  labels = scales::percent) + 
theme_minimal()
```
