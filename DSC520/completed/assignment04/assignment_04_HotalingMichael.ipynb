<br> Load the file `"data/nytimes/covid-19-data/us-states.csv"` and
<br>assign it to the `covid_df` dataframe


```r
covid_df <- read.csv("data/nytimes/covid-19-data/us-states.csv")
head(covid_df)
```

Parse the date column using `as.Date()`

```r
covid_df$date <- as.Date(covid_df$date)
head(covid_df)
```

Create three dataframes named `california_df`, `ny_df`, and `florida_df`
<br>containing the data from California, New York, and Florida

```r
california_df <- subset(covid_df, state == "California")
ny_df <- subset(covid_df, state == "New York")
florida_df <- subset(covid_df, state == "Florida")
```

Plot the number of cases in Florida using `geom_line()`

```r
ggplot(data=florida_df, aes(x=date, y=cases, group=1)) + geom_line()
```

Add lines for New York and California to the plot


```r
ggplot(data=florida_df, aes(x=date, group=1)) +
  geom_line(aes(y = cases)) +
  geom_line(data=ny_df, aes(y = cases)) +
  geom_line(data=california_df, aes(y = cases))
```


Use the colors "darkred", "darkgreen", and "steelblue" for Florida, New York, and California


```r
ggplot(data=florida_df, aes(x=date, group=1)) +
  geom_line(aes(y = cases), color = "darkred", size = 1) +
  geom_line(data=ny_df, aes(y = cases), color= "darkgreen", size = 1) +
  geom_line(data=california_df, aes(y = cases), color= "steelblue", size = 1)
```

Add a legend to the plot using `scale_colour_manual`
<br>Add a blank (" ") label to the x-axis and the label "Cases" to the y axis

```r

ggplot(data=florida_df, aes(x=date, group=1)) +
  geom_line(aes(y = cases, colour = "Florida"), size = 1) +
  geom_line(data=ny_df, aes(y = cases,colour="New York"), size = 1) +
  geom_line(data=california_df, aes(y = cases, colour="California"), size = 1) +
  scale_colour_manual("",
                      breaks = c("Florida", "New York", "California"),
                      values = c("darkred", "darkgreen", "Steelblue")) +
  xlab(" ") + ylab("Cases")
```

Scale the y axis using `scale_y_log10()`

```r

ggplot(data=florida_df, aes(x=date, group=1)) +
  geom_line(aes(y = cases, colour = "Florida"), size = 1) +
  geom_line(data=ny_df, aes(y = cases,colour="New York"), size = 1) +
  geom_line(data=california_df, aes(y = cases, colour="California") , size = 1) +
  scale_colour_manual("",
                      breaks = c("Florida", "New York", "California"),
                      values = c("darkred", "darkgreen", "Steelblue")) +
  xlab(" ") + ylab("Cases") + scale_y_log10()
```
