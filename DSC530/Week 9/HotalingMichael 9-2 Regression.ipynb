### Exercise 11.1 

Suppose one of your co-workers is expecting a baby and you are participating in an office pool to predict the date of birth. Assuming that bets are placed during the 30th week of pregnancy, what variables could you use to make the best prediction? You should limit yourself to variables that are known before the birth, and likely to be available to the people in the pool.

```python
import nsfg
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = nsfg.ReadFemPreg()
df = df[df['outcome'] == 1]
df = df[df['prglngth'] > 30]
```

I chose the attributes:<br>
`feelinpg`: How happy the mother is to have the child <br>
`fmarital`: The formal marital status <br>
`workpreg`: Did the mother work while pregnant <br>
`laborfor`: Which line of work the mother is in<br>
`agecon`: Age of conception<br>
`religion`: The religion of the mother<br>
`metro`: Where the mother lives<br>
`timingok`: How the mother feels about the timing of the pregnancy<br>
`race`: race of the mother<br>
`hieduc`: the educational status of the mother<br>
`hispanic`: if the mother is hispanic or not<br>
`poverty`: poverty level income status<br>
`babysex`: the sex of the baby<br>
`birthord`: which number child this baby will be for the mother<br>
`multbrth`: the number of children born from this pregnancy<br>

```python
attributes = ["feelinpg > 5", "fmarital == 1", "workpreg == 1", "laborfor", 
              "agecon", "C(religion)", "C(metro)", "timingok", 
              "C(race)", "hieduc", "C(hispanic)" ,"poverty",
              "babysex == 1", "birthord == 1", "multbrth == 1"]
```

```python
def model_formulator(target, attribute_list, inter = True):
    str = target + " ~ "
    str += attribute_list[0]
    for i in attribute_list[1:len(attribute_list)]:
        str += " + " + i
    if inter == True:
        str += " + 1"
    return str
```

```python
model = smf.ols(model_formulator("prglngth", attributes), data=df)
results = model.fit()
results.summary()
```

Don't bet the farm on this one. Our R-squared value is very poor. The only realy significant attribute that carries a large amount of weight is if the the coworker will be having multiple babies.

### Exercise 11.3
If the quantity you want to predict is a count, you can use Poisson regression, which is implemented in StatsModels with a function called `poisson`. It works the same way as `ols` and `logit`. As an exercise, let’s use it to predict how many children a woman has born; in the NSFG dataset, this variable is called `numbabes`.

Suppose you meet a woman who is 35 years old, black, and a college graduate whose annual household income exceeds $75,000. How many children would you predict she has born?

```python
import first
import numpy as np

live, firsts, others = first.MakeFrames()

# This part is pretty much entirely lifted from the solution. I had a hard time joining those dataframes. 

resp = nsfg.ReadFemResp()
resp.index = resp.caseid
join = live.join(resp, on='caseid', rsuffix='_r')

join['numbabes'].replace([97], np.nan, inplace=True)
```

```python
attributes = ['age_r', "C(race)", "totincr", "educat"]

model = smf.poisson(model_formulator("numbabes",attributes), data=join)
results = model.fit()
results.summary() 
```

```python
cols = ['age_r', 'race', 'totincr', 'educat']
predict = pd.DataFrame([[35,  1, 14, 16]], columns=cols)
round(results.predict(predict))
```

### Exercise 11.4 
If the quantity you want to predict is categorical, you can use multinomial logistic regression, which is implemented in StatsModels with a function called `mnlogit`. As an exercise, let’s use it to guess whether a woman is married, cohabitating, widowed, divorced, separated, or never married; in the NSFG dataset, marital status is encoded in a variable called `rmarital`.

Suppose you meet a woman who is 25 years old, white, and a high school graduate whose annual household income is about $45,000. What is the probability that she is married, cohabitating, etc?

```python
attributes = ['age_r', "C(race)", "totincr", "educat"]

model = smf.mnlogit(model_formulator("rmarital" , attributes), data=join)
results = model.fit()
results.summary()
```

```python
columns = ['age_r', 'race', 'totincr', 'educat']
new = pd.DataFrame([[25, 2, 11, 12]], columns=columns)
results.predict(new)
```
