```python
import pandas as pd
import numpy as np
import scipy as ss
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_absolute_error, accuracy_score
```

```python
train = pd.read_csv("titanic_train.csv")
test = pd.read_csv("titanic_test.csv")
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train = train.dropna(subset = ["Age"])
test = test.dropna(subset = ["Age"])
test = test.dropna(subset = ["Fare"])
```

```python
train.head()
```

```python
# Need to convert sex to binary

sex_list = []

for i in train['Sex']:
    if i == "male":
        sex_list.append(1)
    else:
        sex_list.append(0)
        
train['Sex_Male'] = sex_list


sex_list = []

for i in test['Sex']:
    if i == "male":
        sex_list.append(1)
    else:
        sex_list.append(0)
        
test['Sex_Male'] = sex_list
```

```python
train = train[["Survived","Age", "Pclass", "Sex_Male", "Fare"]]
test = test[["Age", "Pclass", "Sex_Male", "Fare"]]
```

```python
columns = ["Age", "Pclass", "Sex_Male", "Fare"]
```

```python
x = train[columns]
y = train['Survived']

log_model = LogisticRegression()
```

```python
log_model.fit(x,y);
```

```python
results = log_model.predict(test)
```

```python
test = pd.read_csv("titanic_test.csv")

test = test.dropna(subset = ["Age"])
test = test.dropna(subset = ["Fare"])
test['Predicted Value'] = results
```
