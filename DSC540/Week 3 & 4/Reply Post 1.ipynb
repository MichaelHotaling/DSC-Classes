```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv") # Importing the data
df = df.drop('label',axis = 1) # Dropping the labels to prevent issues with reshaping
df = df/255  # Dividing by 255 to get a value between 0 and 1 for graphing
```

```python
df.head()
```

```python
sample_df = df.sample(n=5, random_state=1234) # Selecting a random value from the dataframe
sample_df
```

```python
size = int(np.sqrt(784)) # Extracting the dimensions of the image
size 
```

```python
for index, row in sample_df.iterrows():
    sample = np.array(row).reshape((size, size)) # Reshaping the image
    plt.imshow(sample, interpolation='nearest', ) # Plotting the image
    plt.gray()
    plt.show()
```

```python
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
  
# Read Images 
img = mpimg.imread('myface.jpg') 
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img);
```

```python
img.shape
```

```python
import matplotlib as mpl
from matplotlib.colors import ListedColormap

def colorFader(c1,c2,mix=0): #fade (linear interpolate) from color c1 (at mix=0) to c2 (mix=1)
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix)*c1 + mix*c2)

c1='#000000' # Max Black
c2='#ff0000' # Max Red
n=255

colors = []
for i in range(0,n):
    colors.append(colorFader(c1,c2,i/n))
    
reds = ListedColormap(colors)
```

```python
plt.imshow(img[::,::,0], cmap=reds); # RED
```

```python
c2='#00ff00' # Max Green
n=255

colors = []
for i in range(0,n):
    colors.append(colorFader(c1,c2,i/n))
    
greens = ListedColormap(colors)

plt.imshow(img[::,::,1], cmap=greens); # GREEN
```

```python
c2='#0000ff' # Max Blue
n=255

colors = []
for i in range(0,n):
    colors.append(colorFader(c1,c2,i/n))
    
blues = ListedColormap(colors)

plt.imshow(img[::,::,2],cmap=blues); # BLUE
```
