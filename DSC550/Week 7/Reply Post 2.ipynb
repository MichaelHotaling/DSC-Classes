```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("phones.csv")
df
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
attributes = df.iloc[:,:-1]
target = df.iloc[:,-1]
```

```python
fit = bestfeatures.fit(attributes,target)
```

```python
dfscores = pd.DataFrame(fit.scores_)
```

```python
dfscores.index = attributes.columns
dfscores.sort_values(by=0, ascending=False)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(attributes,target)
```

```python
tree = pd.DataFrame(model.feature_importances_)
tree.index = attributes.columns
tree.sort_values(by=0, ascending=False)
```
