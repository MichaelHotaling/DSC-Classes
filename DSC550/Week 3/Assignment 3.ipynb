### Using Fake Data

```python
fruits = ['apple','orange','banana','lime', 'blueberry','grapes']
number = [6, 5, 2, 1, 6, 3]
colors = ['red','orange','yellow','lime','blue','purple']
```

```python
plt.figure(figsize=(16,9))
plt.scatter(fruits,
            number, 
            color = colors, 
            s = 1000, 
            zorder=2)
plt.scatter(fruits,
            number, 
            color = 'black', 
            s = 1700,
            zorder=1);
plt.plot(fruits,
         number, 
         c = "grey", 
         linewidth = 10, 
         zorder=0)
plt.ylim(0,
         max(number) + 1);
plt.grid(True)
plt.title("Favorite Fruits")
plt.ylabel("Number of Respondents")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(fruits, 
        number, 
        color = colors, 
        zorder = 2);
plt.bar(fruits, 
        np.array(number)+0.05, 
        color = 'black', 
        width=0.85, 
        zorder=1);
plt.grid(True)
plt.title("Favorite Fruits")
plt.ylabel("Number of Respondents")
plt.show()
```

```python
plt.figure(figsize=(16,16))
plt.pie(number, 
        labels=fruits, 
        colors=colors, 
        autopct = lambda p:f'{p:.1f}%', 
        wedgeprops={"edgecolor":"black",
                    'linewidth': 3,
                    'linestyle': 'solid', 
                    'antialiased': True});
plt.title("Favorite Fruits")
plt.show()
```

### Using Real Data

I'll be using the Toy Dataset from Kaggle for a real set of data. https://www.kaggle.com/carlolepelaars/toy-dataset

```python
df = pd.read_csv("toy_dataset.csv")
```

```python
df.head()
```

```python
df2 = pd.DataFrame(df['City'].value_counts(), columns=(['City']))
df2
```

```python
my_colors = colors = ['red','orange','yellow','green','cyan','blue','purple','pink']
```

```python
plt.figure(figsize=(16,9))
plt.plot(df2.index, 
         df2['City'], 
         color = 'grey',
         linewidth = 5,
         zorder=0)
plt.scatter(x = df2.index, 
            y = df2['City'], 
            color = my_colors, 
            s = 1000, 
            zorder=2,
            edgecolors='black')
plt.ylim(0,max(df2['City']) + (0.10 * max(df2['City'])));
plt.grid(True)
plt.title("Number of Entries for Each City")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.bar(df2.index, df2['City'],color = my_colors,
                                                           edgecolor='b',
                                                           zorder=2)
plt.ylim(0,max(df2['City']) + (0.10 * max(df2['City'])));
plt.grid(True)
plt.title("Number of Entries for Each City")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (16,16))
plt.pie(df2['City'], 
        labels = df2.index, 
        colors = my_colors, 
        autopct = lambda p:f'{p:.1f}%', 
        wedgeprops={"edgecolor":"black",
                    'linewidth': 1,
                    'linestyle': 'solid', 
                    'antialiased': True});
plt.title("Number of Entries for Each City")
plt.show()
```

When visualizing the number of values, a certain category has, a bar graph is the best visualization since it easily shows number of values that each category contains. Bar graphs are the most common method for visualizing this sort of data since each value can be compared with other values.

Pie charts are also useful, but they do not accurately display the number of counts since the values are expressed as a percentage of the pie compared with the other values in the dataset. Pie charts are also considered more difficult to read due to a readers inability to accurately measure the angles and sizes of each slice at a glance. When the values in the data frame are close, it can be almost impossible to distinguish which slice is the largest and which is the smallest. Here is an example to demonstrate this issue.


```python
data1 = [17,18,20,22,24]
data2 = [20,18,21,20,20]
data3 = [24,23,21,19,18]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3);
fig.set_figwidth(15)
ax1.pie(data1, labels = ['a','b','c','d','e']);
ax2.pie(data2, labels = ['a','b','c','d','e']);
ax3.pie(data3, labels = ['a','b','c','d','e']);
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figwidth(15)
ax1.bar(['a','b','c','d','e'],data1);
ax2.bar(['a','b','c','d','e'],data2);
ax3.bar(['a','b','c','d','e'],data3);
```

We can easily see that the bar graph is much easier to understand. We can add values to each slice of the pie chart to display the value of each slice more accurately, but that seems to invalidate the point of using a chart in the first place. 

Pie charts are also only useful when a limited number of categories are present in a data frame, since adding more and more values cause the slices to shrink in size, causing them to be indistinguishable. 



```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3);
fig.set_figwidth(15)
ax1.pie(data1, labels = ['a','b','c','d','e']);
ax2.pie(data2*5, labels = ['a','b','c','d','e']*5);
ax3.pie(data3*10, labels = ['a','b','c','d','e']*10);
```

Pie charts also can't show values that are negative. Bar charts can. This can be important if we want to show values like revenue or other attributes that can be negative.

```python
data1 = [5,8,3,-1,6, -5, ]
plt.pie(data1, labels = ['a','b','c','d','e','f'])
```

```python
data1 = [5,8,3,-1,6, -5, ]
plt.bar(['a','b','c','d','e','f'], data1);
```

Line charts are the worst at displaying the count of categorical data. Line charts imply a continuation of the data, like a time series. Since this type of data has no ordinal values, we should not use a line to demonstrate the count.
Line charts are also not designed for displaying exact values since the y axis doesnt typically start at 0, unlike bar charts. Line charts typically show relative changes within the series, so using like charts to display this sort of data is a very poor decision.
