```python
import numpy as np
import pandas as pd
```

**1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive).**

```python
# Using numpy to generate 500 random numbers into an array of size 10,50
nums = np.random.randint(1,501, size=(10, 50))
```

**Load it to a DataFrame.**

```python
# Loading the array into a dataframe using pandas. Column names will be 1-50
df = pd.DataFrame(nums, columns=np.arange(1,51))

# Forcing the index to start at 1
df.index = np.arange(1, len(df) + 1)
df
```

**2) Calculate: <br>
a) the sum of each row:**

```python
# Creating an iterative loop that sums all 10 rows.

for i,j in enumerate(df.sum(axis=1), start = 1):
    print("Row {:>2} Sum: {}".format(i, j))
```

**b) the sum of each column**

```python
# Creating an iterative loop that sums all 10 columns.

for i,j in enumerate(df.sum(axis=0), start = 1):
    print("Column {:>2} Sum: {}".format(i,j))
```

**c) and the sum of all entries**

```python
# Using df.values.sum() to sum all the values in the dataframe

print("Total Sum: {}".format(df.values.sum()))
```

**3. Determine the minimum; maximum; average; and standard deviation of all entries.**

```python
# Using df.values.max() to find the largest value in the dataframe

df.values.max()
```

```python
# Using df.values.min() to find the smallest value in the dataframe

df.values.min()
```

```python
# Using df.values.mean() to find the mean in the dataframe

df.values.mean()
```

```python
# Using df.values.std() to find the standard deviation in the dataframe

df.values.std()
```

**4) Sort the DataFrame: by rows, by columns, on row 2, on column 2.**

```python
# Creating copies of the first dataframe for the different sorts.

by_row = df.copy()
by_col = df.copy()
by_row_2 = df.copy()
by_col_2 = df.copy()
```

```python
# Sorting each row from lowest value to highest value
# Lowest values start at top of the dataframe and the largest values are at the bottom

by_row.values.sort(axis = 0)
by_row
```

```python
# Sorting each column from lowest value to highest value
# Lowest values start at left of the dataframe and the largest values are at the right

by_col.values.sort(axis = 1)
by_col
```

```python
# Sorting all values in the dataframe based on the contents of row 2. 
# Retains the relative position of all the points while sorting the rows by row 2s contents

by_row_2.sort_values(axis = 0, by = 2)
```

```python
# Sorting all values in the dataframe based on the contents of column 2. 
# Retains the relative position of all the points while sorting the columns by column 2s contents

by_col_2.sort_values(axis = 1, by = 2)
```
