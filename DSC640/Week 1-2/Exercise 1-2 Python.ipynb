# 1.2 Exercises: Charts - Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
hotdog_winners = pd.read_excel("hotdog-contest-winners.xlsm")
hotdog_places = pd.read_excel("hotdog-places.xlsm")
obama_approval_ratings = pd.read_excel("obama-approval-ratings.xls")
```

# Bar Plots

```python
plt.figure(figsize = (16,9))
plt.bar(x=hotdog_winners['Year'], height = hotdog_winners['Dogs eaten'],width = 0.3, color = "red", zorder= 1)
plt.bar(x=hotdog_winners['Year'], height = hotdog_winners['Dogs eaten'],width = 0.7 , color = "gold", zorder = 0)

for year, hotdogs in zip(hotdog_winners['Year'], hotdog_winners['Dogs eaten']):
    plt.plot(np.ones(1000)*year + 0.05 * np.sin(np.linspace(0, hotdogs, 1000)*2), np.linspace(0,hotdogs,1000), linewidth=2.5, color = "yellow", zorder = 2)
plt.ylabel("Number of Hotdogs Eaten")
plt.xlabel("Year")
plt.title("World Record Number of Hotdogs Eaten in One Sitting by Year")
plt.grid(True)
plt.show()
```

```python
hotdog_winners['Winner'] = hotdog_winners['Winner'].str.replace('Takeru "Tsunami" Kobayashi', 'Takeru Kobayashi')
hotdog_winners['Winner'].value_counts(ascending=True).plot(kind='barh', figsize = (16,9), color = "gold")
plt.grid(True)
plt.title("Number of Hotdog Championships Won by Contestant")
plt.show()
```

# Stacked Bar Charts

```python
df1 = hotdog_places.T
df1.columns = ["First Place","Second Place", "Third Place"]
df1 = df1[reversed(df1.columns)]

df1.plot(kind = "bar", stacked=True, figsize = (16,9), color = ['brown','silver','gold'])
plt.title("Total Number of Hot Dogs Eaten between the Top Three Contestants by Year")
plt.xlabel("Year")
plt.ylabel("Number of Hotdogs Eaten")
plt.legend()
plt.show()
```

# Pie Chart

```python
df2 = obama_approval_ratings[obama_approval_ratings.columns[1:]].T
df2.columns = obama_approval_ratings['Issue']

df2['Environment'].plot(kind="pie", autopct='%1.1f%%', startangle=90, figsize = (16,9), colors= ['green','red','grey'])
plt.title("Obama's Approval Rating on Environmental Issues")
plt.ylabel("")
plt.show()
```

# Doughnut Plot

```python
df2['Immigration'].plot(kind="pie", 
                        autopct='%1.1f%%', 
                        startangle=90, 
                        #figsize = (16,9), 
                        colors= ['green','red','grey'],
                        figsize = (16,9))

circle = plt.Circle((0,0), 0.77, color='white')
p = plt.gcf()
p.gca().add_artist(circle)
plt.title("Obama's Approval Rating on Immigration Issues")
plt.ylabel("")
plt.show()
```
