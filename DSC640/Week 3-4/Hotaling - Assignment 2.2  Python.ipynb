# 2.2 Exercises: Line Charts and Step Charts - Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("world-population.xlsm")
```

```python
df.head()
```

# Line Plot

```python
plt.figure(figsize=(16,9))
plt.plot(df['Year'], df['Population'], color = "gray")
plt.scatter(df['Year'], df['Population'], color = "black")
plt.title("World Population by Year")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
```

# Step Plot

```python
plt.figure(figsize=(16,9))
plt.step(df['Year'], df['Population'], color = "grey")
plt.scatter(df['Year'], df['Population'], color = "black")
plt.scatter(df['Year'][:-1], df['Population'][1:], color = "black")
plt.title("World Population by Year")
plt.xlabel("Year")
plt.ylabel("Population")
plt.grid(True)
```
