# 2.2 Exercises: Line Charts and Step Charts - R

```r
library(ggplot2)
library(readxl)
```

```r
df <- read_excel('world-population.xlsm')
head(df)
```

# Line Plot

```r
ggplot(data = df, aes(x = Year, y = Population)) + 
geom_line()+
geom_point() + 
theme_bw() +
ggtitle('World Population by Year')
```

# Step Plot

```r
ggplot() + 
geom_step(data = df, aes(x = Year, y = Population))+
geom_point(data = df, aes(x = Year, y = Population)) + 
geom_point(aes(x = df$Year[-(1)], y = df$Population[-(50)])) + 
theme_bw() +
ggtitle('World Population by Year')
```
